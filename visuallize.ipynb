```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 엑셀 파일 불러오기
file_path = '/content/기타강의.xlsx'  # 실제 파일 경로로 바꿔주세요
data = pd.read_excel(file_path)

# '조회수' 열의 데이터 추출 및 내림차순 정렬
view_counts = data['조회수']
view_counts = view_counts.sort_values(ascending=False)

# 100개씩 구간을 나누어 첫 번째 조회수 대비 백분율 계산
first_view_count = view_counts.iloc[0]  # 첫 번째 조회수

percentages = []
for i in range(0, len(view_counts), 100):
    subset = view_counts.iloc[i:i+100]
    percentage = (subset.iloc[-1] / first_view_count) * 100  # 첫 번째 조회수 대비 백분율 계산
    percentages.append(percentage)

# 그래프로 시각화
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, len(percentages) + 1), percentages, marker='o', linestyle='-')
plt.xlabel('강의 구간 (100개 단위)')
plt.ylabel('첫 번째 조회수 대비 백분율')
plt.title('100개 단위 조회수 구간별 백분율')
plt.xticks(np.arange(1, len(percentages)))  # xticks를 1부터 시작하도록 설정
plt.grid(True)
plt.show()
```
